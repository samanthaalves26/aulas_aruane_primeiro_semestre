```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph()

G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")

G.add_edge("A", "B")
G.add_edge("B", "C")
G.add_edge("C", "D")
G.add_edge("D", "A")

cores = ["red", "green", "blue", "orange"]  # mesma ordem dos nós em G.nodes()

# Desenhar o grafo
nx.draw(
    G,
    with_labels=True,
    node_color=cores,
    node_size=1500,
    font_size=14,
    arrows=True
)
plt.title("Exemplo de Grafo Direcionado")
plt.show()
```
